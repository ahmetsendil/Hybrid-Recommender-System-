```python
# Encapsulate all operations in functions
def prepare_data(movie_path, rating_path):
    
    movie = pd.read_csv("/kaggle/input/movie-and-rating-datasets/movie.csv")
    rating = pd.read_csv("/kaggle/input/movie-and-rating-datasets/rating.csv")

   
    merged_df = pd.merge(rating, movie, on='movieId')

    
    movie_counts = merged_df['title'].value_counts()
    popular_movies = movie_counts[movie_counts >= 1000].index
    filtered_df = merged_df[merged_df['title'].isin(popular_movies)]

    
    pivot_df = filtered_df.pivot_table(index='userId', columns='title', values='rating')

    return pivot_df
```

```python
# Prepare the data using the function
movie_file_path = "/kaggle/input/movie-and-rating-datasets/movie.csv"
rating_file_path = "/kaggle/input/movie-and-rating-datasets/rating.csv"
pivot_table = prepare_data(movie_file_path, rating_file_path)
print(pivot_table.head())
```
